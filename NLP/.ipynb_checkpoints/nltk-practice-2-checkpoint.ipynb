## 2.8.8

```python
import nltk
import matplotlib
```

```python
from nltk.corpus import names
```

```python
cfd = nltk.ConditionalFreqDist(
    (fileid, name[0])
    for fileid in names.fileids()
    for name in names.words(fileid)
)
cfd.plot()
```

## 2.8.11

```python
import nltk
from nltk.corpus import brown
```

```python
cfd2 = nltk.ConditionalFreqDist(
        (genre, word)
        for genre in brown.categories()
        for word in brown.words(categories=genre)
)
word = ['love','like','begin','start']
genres = ['news','romance']
cfd2.tabulate(conditoins=genres,samples=word)
```

## 2.8.14

```python
import nltk
from nltk.corpus import wordnet as wn
def supergloss(s):
    s_definition = s.definition()
    s_hypers = s.hypernyms()
    s_hypos = s.hyponyms()
    s_hypers_definitions = ''
    s_hypos_definitions = ''
    for hyper in s_hypers:
        s_hypers_definitions += hyper.definition() + ' '
    for hypo in s_hypos:
        s_hypos_definitions += hypo.definition() + ' '
    return 's_definition:',s_definition,'s_hypers_definitions',s_hypers_definitions,'s_hypos_definitions',s_hypos_definitions
        
print(supergloss(wn.synset('car.n.01')))  
```
